# Task 1.3 Implementing Laplace Smoothning

```python
def laplace_smoothning_unigram(ngram,corpus):
	unique_corpus = []
	for num in corpus:
		if num not in unique_corpus:
			unique_corpus.append(num)	
	unigram_count =0
	occur = 0
	for i in range(0,len(corpus)):
		if(ngram == corpus[i]):
			unigram_count += 1
			occur = 1
	# print(unigram_count+1)
	# print(len(unique_corpus))
	# print(len(corpus))
	if(occur):
		return float(unigram_count+1)/float(len(corpus)+len(unique_corpus))		
	else:
		return float(1)/float(len(corpus)+len(unique_corpus))
```

code to find the wrong spelling in a sentence and then finding the
minimum distance of that word from all respective unigrams of the corpus
and printing out the most probable spelling(using dictionary method)

```python
memo = {}

def calculate_distance(word1, word2):
    if word1 == "":
        return len(word1)
    if word2 == "":
        return len(word2)
    cost = 0 if word1[-1] == word2[-1] else 1
       
    i1 = (word1[:-1], word2)
    if not i1 in memo:
        memo[i1] = calculate_distance(*i1)
    i2 = (word1, word2[:-1])
    if not i2 in memo:
        memo[i2] = calculate_distance(*i2)
    i3 = (word1[:-1], word2[:-1])
    if not i3 in memo:
        memo[i3] = calculate_distance(*i3)
    res = min([memo[i1]+1, memo[i2]+1, memo[i3]+cost])
    
    return res	
correct_unigram = []

def spell_check_dict_method(sentence):
	words = get_words(sentence)
	# print(words)
	possible_spelling = []
	for word in words:
		if word not in unique_corpus:
			wrong_spelling = word
		else:
			correct_unigram.append(word)
	# print(wrong_spelling)
	best_replace = None
	min_distance = math.inf
	for current in unique_corpus:

		distance = calculate_distance(wrong_spelling,current)
		if 0<distance<=min_distance:
			min_distance = distance
			best_replace = current
			possible_spelling.append(best_replace)
	# print(possible_spelling)
	possible_spelling2 = []
	possible_spelling3 = []
	possible_spelling4 = []
	for current in possible_spelling:
		temp = calculate_distance(current,wrong_spelling)
		if(temp==min_distance):
			possible_spelling2.append(current)

	for current in possible_spelling2:
		if len(current) != 1:
			possible_spelling3.append(current)

	for current in possible_spelling3:
		if (abs(len(current)-len(wrong_spelling))>=0 and abs(len(current)-len(wrong_spelling)) <= 2):
			possible_spelling4.append(current)

	return possible_spelling4
```
